```python
def asterix_star(height):
    count=0
    stars=1
    base=height*2-1
    while count<height:
        str=""
        for i in range(0,base+1):
            if i<height-count or i>height+count:
                str+=" "
            else:
                str+="*"
        count+=1
        print(str)
```
