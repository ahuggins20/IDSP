```python
def TicTacToe():
    times=int(input("Best of ?: "))
    needWins=1+times//2
    wins=[0,0]
    games=0
    while wins[0]<needWins and wins[1]<needWins:
        turns=0
        print("Starting New Game...\n")
        board=[[" "," "," "],[" "," "," "],[" "," "," "]]
        turn=False #false is p1 turn, true is p2 turn
        print_Board(board)
        while check_Win(board)==False and turns<9:
            if turn==False:
                print("Player 1's turn:")
            else:
                print("Player 2's turn:")
            strIn=input("Enter input: ex. (row(1,2,3),column(1,2,3)): ")
            list=strIn.split(",")
            play=(int(list[0])-1, int(list[1])-1)
            if play[0]>2 or play[0]<0 or play[1]>2 or play[1]<0:
                print("Invalid Entry")
                continue
            if board[play[0]][play[1]]==" ":
                if turn==False:
                    board[play[0]][play[1]]="X"
                    turns+=1
                else:
                    board[play[0]][play[1]]="O"
                    turns+=1
                if check_Win(board)==True:
                    break
                else:
                    turn= not turn
            else:
                print("That space is taken")
            print_Board(board)
        if check_Win(board)==True and turn==False:
            wins[0]+=1
            games+=1
            print("Player 1 wins game "+str(games))
        elif check_Win(board)==True and turn==True:
            wins[1]+=1
            games+=1
            print("Player 2 wins game "+str(games))
        elif turns==9 and check_Win(board)==False:
            print("The game is a tie")
        print("The score is "+str(wins[0])+" to "+str(wins[1]))
    if wins[0]==needWins:
        print("Player 1 wins the match!")
    if wins[1]==needWins:
        print("Player 2 wins the match")
        
def check_Win(board):
    if board[0][0] is board[1][1] and board[0][0] is board[2][2] and board[0][0] is not " ":
            return True
    if board[2][0] is board[1][1] and board[2][0] is board[0][2] and board[2][0] is not" ":
            return True
    for x in range(0, len(board)):
        if board[x][0] is not" " and board[x][0] is board[x][1] and board[x][0] is board[x][2]:
            return True
        if board[0][x] is not" " and board[0][x] is board[1][x] and board[0][x] is board[2][x]:
            return True
    return False

def print_Board(board):
    print()
    print(*board[0], sep="|")
    print("_ _ _")
    print(*board[1], sep="|")
    print("_ _ _")
    print(*board[2], sep="|")
    print()

TicTacToe()
```
